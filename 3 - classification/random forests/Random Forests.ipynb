```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("bank-additional-full.csv", delimiter=";")
```

```python
dataset.head()
```

```python
encoder = LabelEncoder()
scaler = StandardScaler()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X.job = encoder.fit_transform(X.job)
```

```python
X.marital = encoder.fit_transform(X.marital)
```

```python
encode_list = ["education", "default", "housing", "loan", "contact", "month", "day_of_week", "duration", "poutcome"]
for i in encode_list:
    X[i] = encoder.fit_transform(X[i])
```

```python
X.iloc[:, 10:11] = scaler.fit_transform(X.iloc[:, 10:11])
```

```python
X.iloc[:, 12:13] = scaler.fit_transform(X.iloc[:, 12:13])
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
